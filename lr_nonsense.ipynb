The actual algorithm

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def loss(x, y, w):
    margin = np.dot(x, w)
    
    return y * -np.logaddexp(0, np.exp(margin) + (1 - y) * (1 + np.logaddexp(0, np.exp(margin))))
```

```python
def loss_grad(x, y, w):
    pred = sigmoid(np.dot(x, w))
    loss = pred - y
    
    grad = np.empty(w.shape)
    
    for i in range(len(w)):
        grad[i] = loss * x[i]
    
    return grad
```

```python
def grad_desc(X, Y, batchsize=5, eta=0.01, num_iter=100):
    
    ## List of thetas at each iteration
    theta_vec = [np.zeros(X.shape[1])]

    for _ in range(num_iter): ## There should be a condition to check if descent is finished
        ## Pick m points
        m_obs = np.random.randint(X.shape[0], size=batchsize)

        ## Calculate gradient for each observation
        grad = np.zeros(X.shape[1])
        for m in m_obs:
            grad += loss_grad(X[m], Y[m], theta_vec[-1])
        grad /= batchsize

        ## Update theta
        theta_vec.append(theta_vec[-1] - (eta*grad))

    return theta_vec[-1]
```
